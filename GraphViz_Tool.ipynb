Viewing dot in xdot format

```python
import subprocess
p1 = subprocess.Popen(["dot", "-Txdot", "Test.dot"], stdout=subprocess.PIPE)
output,err = p1.communicate()
tmp = output.decode()
xdot = tmp.split('\n')
print(xdot)
for line in xdot:
    print(f'>{line}<')
print(err)
```

```python
import pyparsing as pp

greet = pp.Word(pp.alphanums) + "," + pp.Word(pp.alphas) + "!"
for greeting_str in [
            "Hello1, World!",
            "Bonjour, Monde!",
            "Hola, Mundo!",
            "Hallo, Welt!",
        ]:
    greeting = greet.parse_string(greeting_str)
    print(greeting)
```
